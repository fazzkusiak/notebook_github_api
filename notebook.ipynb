```python
# necessary variables, imports

USERNAME = "KentBeck"
ACCESS_KEY = ""

# there was a mention about setting variables to "", however I left them with values
import requests
!pip3 install pandas

# necessary to use pandas in jupyter

import sys
import json
import pandas as pd
```

```python
# github API usage

data = []

headers = {'Authorization': 'token ' + ACCESS_KEY}
api_url = "https://api.github.com/users/" + USERNAME + "/repos"

# collects repositories list

get_repos_list = requests.get(api_url, headers=headers)


```

```python
# loops over every repo
for i in range(len(get_repos_list.json())):
    json_data = get_repos_list.json()[i]
    
    protected_result = "---"
    name = json_data['name']
    size = json_data['size']
    default_branch = json_data['default_branch']
    
    # collects branches list from chosen repo
    
    get_branches_list = requests.get("https://api.github.com/repos/" + USERNAME + "/" + name + "/branches",headers=headers)
    
    # checks if default branch exists and if it is checks if its protected, if it is found we can break from loop
    
    try:
        for j in range(len(get_branches_list.json())):
            if (get_branches_list.json()[j]['name'] == get_repos_list.json()[i]['default_branch']):
                protected_result = get_branches_list.json()[j]['protected']
                break
    except IndexError:
        pass
    
        # checks for contributors count, in case of [] result (no contributors) we catch exception and assign value = 0

    try:
        check_contr_req = len(requests.get("https://api.github.com/repos/KentBeck/" + json_data['name'] + "/contributors?per_page=100",headers=headers).json())
    except json.decoder.JSONDecodeError:
        check_contr_req = 0
        
    # adds found data to array
    
    data.append([name, size, check_contr_req, default_branch, protected_result])
    

```

```python
# using pandas to get tables

dataset = pd.DataFrame(data)

# setting column names

dataset.columns = ["name", "size", "contributors_count", "branch", "is_protected"]

# displaying all rows

pd.set_option('display.max_rows', dataset.shape[0]+1)

# displaying

display(dataset)
```
